```python
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
df["Year"]=df.Start_Time.dt.year
df["Month"]= df.Start_Time.dt.month
df["Hour"] = df.Start_Time.dt.hour
```

```python
sns.distplot(df.Hour,bins=24,kde=False, norm_hist=True)
```

```python
sns.distplot(df.Start_Time.dt.dayofweek,kde=False, norm_hist=True)
```

```python
sunday=df.Start_Time[df.Start_Time.dt.dayofweek == 6]
```

```python
sns.distplot(sunday.dt.hour,kde=False, norm_hist=True)
```

```python
monday=df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday.dt.hour,kde=False, norm_hist=True)
```
